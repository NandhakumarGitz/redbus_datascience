```python
import pandas as pd

# List of CSV file paths
csv_files = [
    "ap_bus_details.csv", "gujarat_bus_details.csv", "himachal_bus_details.csv", "jk_bus_details.csv",
    "kaac_bus_details.csv", "kadamba_bus_details.csv", "kerala_bus_details.csv", "punjab_bus_details.csv",
    "sb_bus_details.csv", "telangana_bus_details.csv", "up_bus_details.csv", "wb_bus_details.csv"
]


df_list = [pd.read_csv(file) for file in csv_files]
df = pd.concat(df_list, ignore_index=True)

# Save the combined DataFrame
df.to_csv("bus_routes.csv", index=False)

# Reload the saved file
df = pd.read_csv("bus_routes.csv")

# Add an 'id' column
df.insert(0, 'id', range(1, len(df) + 1))

# Replace "INR " with an empty string in the 'Price' column
df['Price'] = df['Price'].astype(str).str.replace('INR ', '', regex=False)

# Extract numeric values from the 'Seat_Availability' column
df['Seat_Availability'] = df['Seat_Availability'].astype(str).str.extract('(\d+)')

# Save final DataFrame
df.to_csv("bus_routes_cleaned.csv", index=False)

print("Data processing complete. Saved as 'bus_routes_cleaned.csv'.")
```

```python
#python mysql connection
import pymysql
myconnection = pymysql.connect(host='localhost', user='root',passwd='1234',database="redbuses")
df = pd.read_csv("bus_routes_cleaned.csv")
df = df.dropna()
a = ",".join(f"{i} {j}"
for i,j in zip(df.columns,df.dtypes)).replace("float64","float").replace("object","text").replace("int64","int")
table_name = "bus_routes"
myconnection.cursor().execute(f"create table {table_name} ({a})")
for i in range(len(df)):
    myconnection.cursor().execute(f"insert into {table_name} values {tuple(df.iloc[i])}")
    myconnection.commit()
```
